```python
import pandas as pd

file_path = "/Users/ksw00/PycharmProjects/pythonProject/DSL_EDA/202412_202412_연령별인구현황_월간.csv"

# CSV 파일을 데이터프레임으로 읽기
df = pd.read_csv(file_path)
df
```

```python
df = df.drop(columns=["2024년12월_계_총인구수","2024년12월_계_연령구간인구수"])
df
```

```python
df_total = df.iloc[:,:12]
df_total
```

```python

# 결과 저장 리스트
result = []

# 각 행정구역별로 처리
for index, row in df_total.iterrows():
    # 행정구역명 저장
    region = row["행정구역"]
    # 연령대별 인구수 추출 및 정렬
    age_groups = row[1:].sort_values(ascending=False)[:3]
    # 결과 저장
    result.append({"행정구역": region, "상위 3 연령대": age_groups.index.tolist(), "인구수": age_groups.values.tolist()})

# 결과 데이터프레임 생성
result_df = pd.DataFrame(result)

result_df
```
