```python
import time
import pickle
from selenium.webdriver.common.by import By
import pandas as pd
from datetime import datetime

# 제목 데이터 초기화
titles = []  # 새로운 데이터로 초기화
print("새로운 크롤링 시작!")

# 제목 크롤링
try:
    for i, post in enumerate(d):  # 로드된 요소(d)를 순회
        try:
            # 제목만 추출
            title = post.text.strip()  # 요소에서 텍스트(제목)만 가져옴

            # 제목이 공백이 아니고, 리스트에 중복되지 않는 경우 추가
            if title and title not in titles:
                titles.append(title)
                print(f"{i + 1}번째 제목: {title}")

        except Exception as e:
            print(f"{i + 1}번째 제목 처리 중 오류 발생: {e}")
            continue

    # 데이터 저장 (고유 파일 이름 생성)
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")  # 현재 날짜와 시간으로 고유 이름 생성
    save_path_pickle = f"titles_{timestamp}.pkl"
    save_path_csv = f"titles_{timestamp}.csv"

    # Pickle 파일로 저장
    with open(save_path_pickle, "wb") as file:
        pickle.dump(titles, file)
    print(f"크롤링 완료: 총 {len(titles)}개의 제목 저장 (Pickle: {save_path_pickle})")

    # DataFrame 생성 및 CSV 저장
    df = pd.DataFrame({"title": titles})  # 제목 리스트를 DataFrame으로 변환
    df.to_csv("역삼 핫플.csv", index=False, encoding="utf-8-sig")
    print(f"CSV 파일로 저장 완료! (CSV: {save_path_csv})")

except Exception as e:
    print(f"크롤링 중 오류 발생: {e}")
```

```python
import time
import pickle
from selenium.webdriver.common.by import By

# 리뷰 데이터 저장 파일
SAVE_PATH = "reviews.pkl"

# 중단된 지점에서 시작할 수 있도록 데이터 불러오기
try:
    with open(SAVE_PATH, "rb") as file:
        reviews = pickle.load(file)
    print(f"기존 데이터 불러오기 완료: {len(reviews)}개의 리뷰")
except FileNotFoundError:
    reviews = {}
    print("저장된 데이터가 없어 새로 시작합니다.")

# 키워드 정의
keywords = [
    "내돈내산", "내돈 내산", "솔직후기", "솔직 후기",
    "리얼후기", "리얼 후기", "정직리뷰", "정직 리뷰",
    "광고아님", "광고 아님", "100% 내돈"
]

# 이미 처리한 리뷰 수
start_index = len(reviews)

# 리뷰 데이터 크롤링
for i in range(start_index, len(d)):  # 기존 데이터 이후부터 시작
    try:
        revs = driver.find_elements(By.XPATH, '/html/body/div/div[2]/div[2]/div[2]/div/div[2]/div/div/a')
        driver.execute_script("arguments[0].scrollIntoView({block: 'center'});", revs[i])
        revs[i].click()

        texts = driver.find_elements(By.XPATH, '/html/body/div[6]/div[1]/div[4]/div[1]/div/div//p')
        full_t = ''
        time.sleep(3)

        for t in texts:
            text = t.text
            full_t += ' ' + text

        print(f"{i}번째: {full_t}")

        if any(k in full_t for k in keywords):
            label = 0  # 정직후기 -> 0
        else:
            label = 1

        reviews[full_t] = label

        # 데이터 저장
        with open(SAVE_PATH, "wb") as file:
            pickle.dump(reviews, file)

        driver.back()
        time.sleep(5)
    except Exception as e:
        print(f"{i}번째 리뷰 처리 중 오류 발생: {e}")
        break  # 오류 발생 시 중단

import pandas as pd
df = pd.DataFrame(list(reviews.items()), columns=['review', 'label'])
df
df.to_csv('네이버블로그리뷰_강남구놀거리.csv', index=False, encoding='utf-8-sig')
driver.back()
```

```python
import requests
from bs4 import BeautifulSoup
import time
import pandas as pd
from tqdm import tqdm

query_list = [
    # "종로구",
    # "중구",
    # "용산구",
    # "성동구",
    # "광진구",
    # "동대문구",
    # "중랑구",
     #"성북구",
     #"강북구",
     #"도봉구",
     #"노원구",
     #"은평구",
    #"서대문구",
    #"마포구",
    #"양천구",
    #"강서구",
    #"구로구",
    #"금천구",
    #"영등포구",
    #"동작구",
    #"관악구",
    #"서초구",
    "강남구",
    #"송파구",
    #"강동구",
]

page = 150

for idx, query in enumerate(query_list, start=1):
    query_data = []
    print(f"전체 비율 : {idx}/{len(query_list)}")

    for i in tqdm(range(1, page * 10, 10), desc=f"{query} 뉴스 크롤링 중..."):
        response = requests.get(
            f"https://search.naver.com/search.naver?where=news&query={query}&sm=tab_opt&sort=0&photo=3&field=0&pd=5&ds=2024.01.23&de=2025.01.22&docid=&related=0&mynews=1&office_type=0&office_section_code=0&news_office_checked=&nso=so%3Ar%2Cp%3A1y&is_sug_officeid=0&office_category=0&service_area=0&start={i}"
        )
        html = response.text
        soup = BeautifulSoup(html, "html.parser")
        articles = soup.select("div.info_group")

        for article in articles:
            links = article.select("a.info")
            if len(links) >= 2:
                url = links[1].attrs["href"]
                response = requests.get(url, headers={"User-agent": "Mozilla/5.0"})
                html = response.text
                soup = BeautifulSoup(html, "html.parser")

                if "entertain" in response.url or "sports" in response.url:
                    continue

                try:
                    title = soup.select_one("#title_area").get_text().strip()
                    content = soup.select_one("#newsct_article").get_text().strip()

                    query_data.append({"title": title, "content": content, "url": url})
                except AttributeError:
                    continue

        time.sleep(0.3)

    df = pd.DataFrame(query_data)

    if not df.empty:
        df.to_csv(f"{query}_news.csv", index=False, encoding="utf-8-sig")
    else:
        print(f"데이터 없음")
```
