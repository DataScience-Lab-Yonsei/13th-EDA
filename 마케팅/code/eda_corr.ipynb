```python
import pandas as pd

file_path = "/Users/ksw00/PycharmProjects/pythonProject/DSL_EDA/서울 시민의 소비 및 이동 패턴 재현 데이터.csv"

# CSV 파일을 데이터프레임으로 읽기
df = pd.read_csv(file_path)
gm_df = df.iloc[:, 35:55]

base_column = gm_df.iloc[:, 0]  # 기준 열 (두 번째 열)
gm_df.iloc[:, 1:21] = gm_df.iloc[:, 1:21].multiply(base_column, axis=0)
```

```python
# 이미지의 오른쪽 열 이름 리스트
new_column_names = [
    "가맹점합산이용금액","요식_이용비중", "유흥_이용비중", "백화점_이용비중", "대형할인점_이용비중",
    "소형유통점_이용비중", "의류/잡화_이용비중", "스포츠/문화/레저_이용비중",
    "숙박_이용비중", "여행_이용비중", "교통_이용비중", "미용_이용비중",
    "각종생활/서비스_이용비중", "교육/학원_이용비중","의료_이용비중", "가전/가구_이용비중",
    "자동차판매_이용비중", "자동차서비스/용품_이용비중", "주유_이용비중",
    "전자상거래_이용비중"
]

print(len(new_column_names))
# 열 이름 변경
gm_df.columns = new_column_names
gm_df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import rc

# 한글 폰트 설정 (윈도우 예시)
rc('font', family='Malgun Gothic')

# 음수 기호가 깨지는 문제 해결
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 2번째 열부터 20번째 열 선택
selected_columns = gm_df.iloc[:, 1:21]

# 상관관계 계산
correlation_matrix = selected_columns.corr()

# 히트맵 그리기
plt.figure(figsize=(12, 10))  # 히트맵 크기 조정
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', cbar=True)

# 제목 및 시각적 설정
plt.title('Correlation Heatmap', fontsize=16)
plt.xticks(rotation=90)  # x축 라벨 회전
plt.yticks(rotation=0)   # y축 라벨 회전
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(10, 2, figsize=(20, 100))  # 4x5 형태의 서브플롯
axes = axes.flatten()  # 2차원 배열을 1차원으로 변환

# 각 열별 데이터 분포 시각화
for i, column in enumerate(gm_df.columns):
    sns.histplot(gm_df[column], kde=True, bins=30, ax=axes[i], color='blue', alpha=0.7)
    axes[i].set_title(f'{column} - 데이터 분포')
    axes[i].set_xlabel(column)
    axes[i].set_ylabel('빈도')

# 빈 서브플롯 숨기기 (열이 20개보다 적을 경우 대비)
for j in range(len(gm_df.columns), len(axes)):
    axes[j].set_visible(False)

# 전체 그래프 레이아웃 정리
plt.tight_layout()
plt.show()
```
