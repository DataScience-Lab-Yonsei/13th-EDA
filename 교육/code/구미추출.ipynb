```python
file_path = "/content/drive/MyDrive/Colab Notebooks/DSL/EDA2_Education/data/elementary_info.csv"
df_elem = pd.read_csv(file_path, encoding='utf-8-sig')
```

```python
file_path = "/content/drive/MyDrive/Colab Notebooks/DSL/EDA2_Education/data/초등학교_폐교위험.csv"
df_danger = pd.read_csv(file_path, encoding='utf-8-sig')
```
