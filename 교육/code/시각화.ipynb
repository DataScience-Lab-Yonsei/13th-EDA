```python
import pandas as pd
#6243 - 954
```

```python
file_path = "/content/drive/MyDrive/Colab Notebooks/DSL/EDA2_Education/data/초등학교_정보.csv"
df_elem = pd.read_csv(file_path, encoding='utf-8-sig')
file_path = "/content/drive/MyDrive/Colab Notebooks/DSL/EDA2_Education/data/초등학교_폐교위험.csv"
df_danger = pd.read_csv(file_path, encoding='utf-8-sig')
```

```python
df_elem = df_elem[~df_elem['학교ID'].isin(df_danger['학교ID'])]
```

```python
import folium

# 대한민국 중심에 folium 지도 생성
m = folium.Map(location=[36.5, 127.5], zoom_start=7)

# 각 위치에 마커 추가
for index, row in df_elem.iterrows():
    folium.CircleMarker(
        location=[row['위도'], row['경도']],  # 위도, 경도
        popup=row['학교ID'],  # 마커 클릭 시 나타나는 정보
        tooltip=row['학교명'],  # 마우스 올릴 때 표시
        color='blue',
        fill=True,  # 원 내부 채우기
        fill_color='blue',  # 내부 색상
        radius=1
    ).add_to(m)

for idx, row in df_danger.iterrows():
    folium.CircleMarker(
        location=[row['위도'], row['경도']],  # 위도, 경도
        popup=row['학교ID'],  # 마커 클릭 시 나타나는 정보
        tooltip=row['학교명'],  # 마우스 올릴 때 표시
        color='red',
        fill=True,  # 원 내부 채우기
        fill_color='red',  # 내부 색상
        radius=1
    ).add_to(m)

# 지도 저장
m.save("map.html")
```
