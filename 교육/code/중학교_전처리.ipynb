# 출생아수 데이터와 Merge

```python
df_b = pd.read_csv('출생아수_시_군_구.csv')
df_b.head()
```

- 출생아수 기반으로 해당 지역에 입학할 중학생 수 추정

```python
# '시군구별' 열의 공백 제거
df_b['시군구별'] = df_b['시군구별'].str.strip()

# 열 이름을 정수형으로 변환
df_b.columns = [df_b.columns[0]] + [int(col) for col in df_b.columns[1:]]

# 숫자 데이터를 정수형(int)으로 변환
for col in df_b.columns[1:]:
    df_b[col] = pd.to_numeric(df_b[col], errors='coerce').fillna(0).astype(int)

# 출생아수 데이터에서 3개년 합계를 구해 중학교 학생 수 추정 (올바른 계산 방식 적용)
# 2010년부터 2031년까지의 중학교 예상 학생 수 계산
df_ems = pd.DataFrame({'시군구별': df_b['시군구별']})

for year in range(2012, 2037):  # 2012년부터 2036년까지 추정
    required_years = [year - 13, year - 14, year - 15]  # 해당 연도에 중학생이 되는 출생연도
    valid_years = [y for y in required_years if y in df_b.columns]  # 실제 데이터에 존재하는 연도만 사용
    
    # 출생아 수 합산 (해당 연도가 존재하지 않는 경우 0으로 처리)
    df_ems[f'{year}_해당지역예측학생수'] = df_b[valid_years].sum(axis=1, numeric_only=True) if valid_years else 0

# 0이 포함된 행이 있을 경우 마지막 0이 위치한 이후 연도의 첫 번째, 두 번째 셀을 0으로 설정
for idx, row in df_ems.iloc[:, 1:].iterrows():
    zero_indices = row[row == 0].index
    if len(zero_indices) > 0:
        last_zero_idx = zero_indices[-1]
        next_columns = df_ems.columns[df_ems.columns.get_loc(last_zero_idx) + 1:]
        if len(next_columns) > 0:
            df_ems.loc[idx, next_columns[0]] = 0
        if len(next_columns) > 1:
            df_ems.loc[idx, next_columns[1]] = 0

df_ems
```

- 학생수 + 위치 + 예측학생 수 Merge

```python
# '소재지지번주소'가 NA가 아닌 데이터만 사용
df_merged = df_merged.dropna(subset=['소재지지번주소'])

# '소재지지번주소'에서 '시군구별' 값이 포함된 경우 매핑
def find_region(address):
    for region in df_ems['시군구별']:
        if region in address:
            return region
    return None

# 학생 수 데이터에 '출생아수' 정보를 추가
df_merged['시군구별'] = df_merged['소재지지번주소'].apply(find_region)

# 매칭된 데이터와 출생아수 데이터 병합
df_final = df_merged.merge(df_ems, on='시군구별', how='left')

#열 순서 정렬
col_to_move = '시군구별'
target_index = 7 

cols = [col for col in df_final.columns if col != col_to_move]  # 기존 컬럼에서 제거
cols.insert(target_index, col_to_move)  # 원하는 위치에 삽입

df_final = df_final[cols]

# 예측학생 수 정수 변환
df_final.iloc[:, 24:] = df_final.iloc[:, 24:].fillna(0).astype(int)

df_final.head()
```

```python
df_final.to_csv("중학교_데이터셋.csv", index=False, encoding='utf-8-sig')
```
