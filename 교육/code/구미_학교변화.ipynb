```python
df_ele_alive = df_ele[df_ele['label']==0]
df_ele_closed = df_ele[df_ele['label']==1]
df_mid_alive = df_mid[df_mid['label']==0]
df_mid_closed = df_mid[df_mid['label']==1]
df_high_alive = df_high[df_high['label']==0]
df_high_closed = df_high[df_high['label']==1]
```

```python
import folium
import numpy as np
import random
from sklearn.cluster import DBSCAN
from scipy.spatial import distance_matrix
```

```python
# DBSCAN 클러스터링 (eps: 거리 기준, min_samples: 최소 학교 개수)
dbscan = DBSCAN(eps=0.019, min_samples=2)  # 거리 설정은 지도 크기에 맞게 조정
df_ele_closed['Cluster'] = dbscan.fit_predict(df_ele_closed[['위도', '경도']])

# 클러스터별 색상을 랜덤으로 지정 (최대 20개 색상 활용)
cluster_colors = {}
unique_clusters = df_ele_closed['Cluster'].unique()
color_palette = ['purple', 'indigo', 'darkorange', 'maroon', 'darkcyan', 'magenta']

# 클러스터별 고유한 색상을 지정하기 위한 딕셔너리
cluster_colors = {}
unique_clusters = sorted(df_ele_closed['Cluster'].unique())  # 클러스터 ID 정렬

for cluster in unique_clusters:
    if cluster != -1:  # 이상치 클러스터(-1) 제외
        cluster_colors[cluster] = random.choice(color_palette)

# 색상 리스트를 클러스터별로 순차적으로 배정
for i, cluster in enumerate(unique_clusters):
    if cluster != -1:  # 이상치 클러스터(-1) 제외
        cluster_colors[cluster] = color_palette[i % len(color_palette)]  # 색상 리스트를 순환 적용

# 클러스터별 대표학교 선정
representative_schools = []
for cluster in df_ele_closed['Cluster'].unique():
    if cluster == -1:
        continue  # 이상치(-1) 제거

    cluster_schools = df_ele_closed[df_ele_closed['Cluster'] == cluster]

    # 1️. 해당 클러스터 내에서 생존학교와 가장 가까운 폐교된 학교 찾기
    distances = distance_matrix(cluster_schools[['위도', '경도']],
                                df_ele_alive[['위도', '경도']])

    if distances.size > 0:
        min_index = np.argmin(distances) % len(cluster_schools)
        representative_schools.append(cluster_schools.iloc[min_index])
    else:
        # 2️. 생존학교가 근처에 없으면 클러스터 중심을 대표학교로 설정
        center = cluster_schools[['위도', '경도']].mean()
        representative_schools.append(center)

# 대표학교 리스트를 DataFrame으로 변환
df_representative = pd.DataFrame(representative_schools)

# 지도 생성
m = folium.Map(location=[36.5, 127.5], zoom_start=8)

# 기존 생존학교(파란색 점) 표시
for _, row in df_ele_alive.iterrows():
    folium.CircleMarker(
        location=[row['위도'], row['경도']],
        color='blue', fill=True, fill_color='blue', radius=4
    ).add_to(m)

# 기존 폐교될 학교(클러스트별색 점) 표시
for _, row in df_ele_closed.iterrows():
    cluster = row['Cluster']
    if cluster == -1:
      color = 'black'  # 이상치는 검은색으로 표시
    else:
      color = cluster_colors.get(cluster, 'red')  # 클러스터별 색상 적용

    folium.CircleMarker(
        location=[row['위도'], row['경도']],
        color=color, fill=True, fill_color=color, radius=4
    ).add_to(m)



m.save("gumi_ele_without_integration.html")
```
