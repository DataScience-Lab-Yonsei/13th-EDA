```python
import pandas as pd
import folium
```

```python
df_danger = pd.read_csv('중학교_폐교위험.csv')
df_all = pd.read_csv('중학교_데이터셋.csv')

# 위험 목록에 있는 id인지 확인하여 라벨링
df_all['label'] = df_all['개방ID'].isin(df_danger['개방ID']).astype(int)  # 1: 포함, 0: 미포함
```

```python
result = df_all.groupby("시군구별")["label"].mean().reset_index()
```

```python
result.columns = ["시군구", "폐교율"]
result = result.sort_values(by="폐교율", ascending=False)
result
```

```python
df_location = df_all[['위도','경도','label']]

# 대한민국 중심 좌표 설정
m = folium.Map(location=[36.5, 127.5], zoom_start=7)

# 색상 매핑
color_map = {1: 'red', 0: 'blue'}

# CircleMarker로 점 추가
for _, row in df_location.iterrows():
    folium.CircleMarker(
        location=[row['위도'], row['경도']],
        radius=0.5,
        color=color_map[row['label']],
        fill=True,
        fill_color=color_map[row['label']],
        fill_opacity=0.7
    ).add_to(m)

# 지도 출력
m
```

```python
m.save("middel_school_map.html")
```
