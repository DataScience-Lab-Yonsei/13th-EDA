```python
plt.figure(figsize=(8, 5))

sns.boxplot(data=df, x='CHNNEL_NM', y='TVGRP_CO', palette='Set2')


plt.title("채널별 TVGRP_CO 분포", fontsize=14)
plt.xlabel("CHNNES_NM", fontsize=12)
plt.ylabel("TVGRP_CO", fontsize=12)

# 그래프 출력
plt.show()
```

```python
df
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x=df['ADVRTS_TIME'], order=df['ADVRTS_TIME'].value_counts().index)
plt.xticks(rotation=45)  # x축 글자 회전
plt.xlabel('ADVRTS_TIME')
plt.ylabel('Count')
plt.title('ADVRTS_TIME별 데이터 개수')
plt.show()
```

```python
plt.figure(figsize=(15, 5))

sns.boxplot(data=df, x='ADVRTS_TIME', y='TVGRP_CO', palette='Set2')


plt.title("ADVRTS_TIME별 TVGRP_CO 분포", fontsize=14)
plt.xlabel("ADVRTS_TIME", fontsize=12)
plt.ylabel("TVGRP_CO", fontsize=12)

# 그래프 출력
plt.show()
```

```python
time_grp_avg = df.groupby('ADVRTS_TIME')['TVGRP_CO'].mean().reset_index()

plt.figure(figsize=(12,6))
sns.barplot(x='ADVRTS_TIME', y='TVGRP_CO', data=time_grp_avg, order=time_grp_avg.sort_values('TVGRP_CO', ascending=False)['ADVRTS_TIME'])
plt.xticks(rotation=45)
plt.xlabel('ADVRTS_TIME')
plt.ylabel('TVGRP_CO (평균)')
plt.title('ADVRTS_TIME별 TVGRP_CO 평균')
plt.show()
```

```python
median_grp = df.groupby('ADVRTS_TIME')['TVGRP_CO'].median().reset_index()

# 시각화
plt.figure(figsize=(10, 6))
sns.barplot(data=median_grp, x='ADVRTS_TIME', y='TVGRP_CO', palette='viridis')

# 그래프 제목 및 라벨
plt.title('광고 시간별 GRP 중앙값', fontsize=14)
plt.xlabel('광고 시간 (초)', fontsize=12)
plt.ylabel('GRP 중앙값', fontsize=12)
plt.xticks(rotation=45)  # x축 가독성을 위해 회전
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
