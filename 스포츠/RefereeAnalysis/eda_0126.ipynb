```python
import pandas as pd
import matplotlib.pyplot as plt

# 예제 데이터 (소트니코바 심판별 점수)
data = {
    "Judge": ["J1", "J2", "J3", "J4", "J5", "J6", "J7", "J8", "J9"],
    "TES": [74.33, 79.53, 70.33, 74.83, 77.63, 73.63, 80.73, 71.03, 73.33],
    "PCS": [76.80, 76.40, 72.80, 70.80, 76.40, 74.40, 77.20, 70.80, 72.00]
}
df = pd.DataFrame(data)

# TES와 PCS의 평균과 표준편차 계산
tes_mean = df["TES"].mean()
tes_std = df["TES"].std()
pcs_mean = df["PCS"].mean()
pcs_std = df["PCS"].std()

# 결과 출력
print(f"TES Mean: {tes_mean}, TES Std Dev: {tes_std}")
print(f"PCS Mean: {pcs_mean}, PCS Std Dev: {pcs_std}")

# 시각화: TES와 PCS 점수 비교
plt.figure(figsize=(12, 6))

# TES 점수
plt.subplot(1, 2, 1)
plt.bar(df["Judge"], df["TES"], color="blue")
plt.axhline(y=tes_mean, color="red", linestyle="--", label="TES Mean")
plt.title("TES Scores by Judge (Sotnikova)")
plt.ylabel("TES")
plt.legend()

# PCS 점수
plt.subplot(1, 2, 2)
plt.bar(df["Judge"], df["PCS"], color="green")
plt.axhline(y=pcs_mean, color="red", linestyle="--", label="PCS Mean")
plt.title("PCS Scores by Judge (Sotnikova)")
plt.ylabel("PCS")
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 데이터 예시 (심판별 TES, PCS 점수)
data = {
    "Judge": ["J1", "J2", "J3", "J4", "J5", "J6", "J7", "J8", "J9"],
    "Asada_TES": [68.52, 70.83, 69.12, 71.33, 70.10, 71.50, 68.80, 69.00, 69.90],
    "Asada_PCS": [70.75, 69.90, 71.20, 70.60, 71.10, 71.40, 70.00, 71.20, 70.90],
    "Kostner_TES": [67.40, 68.50, 68.00, 67.80, 68.60, 69.00, 67.20, 67.50, 68.10],
    "Kostner_PCS": [72.30, 71.80, 72.50, 71.90, 72.00, 72.20, 71.60, 72.10, 71.70],
}

# 데이터프레임 생성
df = pd.DataFrame(data)

# 선수 이름들
skaters = ["Asada", "Kostner"]
categories = ["TES", "PCS"]

# TES와 PCS 시각화
for skater in skaters:
    for category in categories:
        scores = df[f"{skater}_{category}"]
        mean_score = np.mean(scores)
        std_dev = np.std(scores)

        # 평균과 표준편차 출력
        print(f"{skater} {category} Mean: {mean_score:.2f}, Std Dev: {std_dev:.2f}")

        # 그래프 생성
        plt.figure(figsize=(12, 6))
        plt.bar(df["Judge"], scores, color="blue" if skater == "Asada" else "green", alpha=0.6, label=f"{skater} {category}")
        plt.axhline(y=mean_score, color="red", linestyle="--", label=f"{skater} {category} Mean")
        plt.title(f"{category} Scores by Judge ({skater})")
        plt.ylabel(category)
        plt.xlabel("Judges")
        plt.legend()
        plt.tight_layout()
        plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 설정
skaters = ["Sotnikova", "Kim", "Asada", "Kostner", "Gold", "Lipnitskaia"]
tes_means = [75.041, 71.417, 69.90, 68.01, 69.03, 65.57]
tes_stds = [3.588, 2.515, 1.06, 0.57, 0.45, 0.58]
pcs_means = [75.111, 72.978, 70.78, 72.01, 69.87, 70.87]
pcs_stds = [2.348, 2.697, 0.50, 0.28, 0.72, 2.36]

x = np.arange(len(skaters))  # x축 위치
width = 0.35  # 바 폭

# 그래프 생성
plt.figure(figsize=(12, 8))

# TES 그래프
plt.bar(x - width/2, tes_means, width, yerr=tes_stds, label="TES", color="blue", alpha=0.7, capsize=5)

# PCS 그래프
plt.bar(x + width/2, pcs_means, width, yerr=pcs_stds, label="PCS", color="green", alpha=0.7, capsize=5)

# 그래프 설정
plt.xticks(x, skaters)
plt.ylabel("Scores")
plt.title("TES and PCS Comparison with Standard Deviations")
plt.legend()
plt.tight_layout()

# 그래프 출력
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 데이터 설정
data = {
    "Skater": ["Sotnikova", "Kim", "Asada", "Kostner", "Gold", "Lipnitskaia"],
    "TES Mean": [75.041, 71.417, 69.90, 68.01, 69.03, 65.57],
    "TES Std Dev": [3.588, 2.515, 1.06, 0.57, 0.45, 0.58],
    "PCS Mean": [75.111, 72.978, 70.78, 72.01, 69.87, 70.87],
    "PCS Std Dev": [2.348, 2.697, 0.50, 0.28, 0.72, 2.36],
}

df = pd.DataFrame(data)

# 데이터프레임 차트로 표시
plt.figure(figsize=(12, 6))
plt.axis('tight')
plt.axis('off')
table = plt.table(cellText=df.values, colLabels=df.columns, loc='center', cellLoc='center')
table.auto_set_font_size(False)
table.set_fontsize(12)
table.auto_set_column_width(col=list(range(len(df.columns))))

plt.title("TES and PCS Mean & Std Dev Comparison", fontsize=14)
plt.show()
```
